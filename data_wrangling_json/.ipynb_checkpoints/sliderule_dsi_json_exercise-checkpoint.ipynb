## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

### Question 2: Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
# there are nested values within this field
wb['mjtheme_namecode'].head()
```

```python
# so I need to try a different load strategy instead. 
%time
themes = json.load((open('data/world_bank_projects.json')))
themes = json_normalize(themes, 'mjtheme_namecode')
themes.head()
```

```python
# 10 most frequent project themes (including blank values addressed in Question 3)
themes.name.value_counts().head(10)
```

****
### Question 3: In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

```python
# create lookup table 
lookup = themes.groupby('code').max().sort_index()
lookup.head()
```

```python
# join back lookup labels on code as the key field
themes_fill = themes.join(lookup, on='code', rsuffix='_clean')

#check a sample to ensure that the join worked as expected
themes_fill.head(20)
```

```python
# clean up by removing the old field and then renaming the new field
themes_fill = themes_fill.drop('name', 1).rename(columns={'name_clean': 'name'})
themes_fill.head()
```
